```python
class DPC(object):
    def __init__(self, data:np.ndarray) -> None:
        self.X = data
        self.m, self.n = self.X.shape
        
        
        X2 = np.square(self.X)
        one = np.ones((self.m, self.n))
        # 利用矩阵运算计算每个元素之间的距离
        self.D = X2 @ one.T + one @ X2.T - 2*self.X @ self.X.T + 100*np.identity(self.m)
        
    
    def calculate(self, dc = 1, rho_rate = 10, delta_rate = 10):
        self.rho_rate = 100 - rho_rate
        self.delta_rate = 100 - delta_rate
        self.rho = np.zeros((self.m, 1))
        self._rho = np.zeros((self.m, 1))
        self.delta = np.zeros((self.m, 1))
        
        for i in range(0, self.m):
            index = np.where((self.D[i:i+1, :] < dc).all(axis=0))[0]
            self.rho[i] = len(index) - 1
            self._rho[i] = self.rho[i] + np.random.rand()/1e4
        
        max_rho = self._rho.max()
        
        for i in range(0, self.m):
            if self._rho[i] == max_rho:
                self.delta[i] = np.max(self.D[i][:])
            else:
                index = np.where((self._rho[:,:] > self._rho[i,0]).all(axis=1))[0]
                self.delta[i] = np.min(self.D[i][index])
                
    
    def _find_center(self):
        rho_index = np.where((self.rho[:, :] > np.percentile(self.rho, self.rho_rate)).all(axis=1))[0]
        delta_index = np.where((self.delta[rho_index, :] > np.percentile(self.delta[rho_index, :], self.delta_rate)).all(axis=1))[0]
        self.center_index = np.array(rho_index)[delta_index]

        all = list(range(0, self.m))
        self.other = []
        for i in all:
            if i not in self.center_index:
                self.other.append(i)


    def _get_cluster(self):
        self.clusters = []
        self.label = np.zeros((self.m, 1))
        
        for i in range(0, len(self.center_index)):
            self.clusters.append([])
            
        for i in self.other:
            i_index = int(np.argmin(self.D[i][self.center_index]))
            self.clusters[i_index].append(i)
            self.label[i] = i_index
        
        for i in range(0, len(self.center_index)):
            self.label[self.center_index[i]] = i 

                 
                
    def draw_decision_graph(self):
        self._find_center()
        self.colors = random.sample(color_list, len(self.center_index))

        

        fig = plt.figure() 
        ax1 = fig.add_subplot(111)
        ax1.set_title('Decision graph') 
        plt.xlabel('Rho') 
        plt.ylabel('Delta') 
        ax1.scatter(self.rho[self.other], self.delta[self.other], c=BLACK, marker='o') 
        for i in range(0, len(self.center_index)):
            ax1.scatter(self.rho[self.center_index[i]], self.delta[self.center_index[i]], c=self.colors[i], marker='o') 
        plt.show() 
        
    
    def draw_cluster(self):
        self._get_cluster()
        fig = plt.figure() 
        ax1 = fig.add_subplot(111)
        ax1.set_title('Cluster graph') 
        plt.xlabel('X') 
        plt.ylabel('Y') 
        
        for i in range(0, len(self.clusters)):
            ax1.scatter(self.X[self.clusters[i], 0], self.X[self.clusters[i], 1], c=self.colors[i], marker='o') 
        ax1.scatter(self.X[self.center_index, 0], self.X[self.center_index, 1], c=BLACK, marker='x')
        plt.show() 
```

## Clustering on different datasets

```python
import sklearn.metrics as sklm

def find_best(data: np.ndarray):
    model = DPC(data)
    
    min = 1000
    min_i = -1
    min_j = -1
    
    for i in range(11,30):
        for j in range(7,30):
            model.calculate(0.2, i, j)
            model._find_center()
            model._get_cluster()
            score = sklm.davies_bouldin_score(data, model.label.flatten())
            if score < min:
                min = score
                min_i = i
                min_j = j

    print(min_i, min_j)
    model.calculate(0.2, min_i, min_j)
    model._find_center()
    model._get_cluster()
    score = sklm.davies_bouldin_score(data, model.label.flatten())
    print("DBI = {}".format(score))
```

```python
data = pd.read_csv('./Datasets/Aggregation.txt',names=['x','y'], header=None,sep=" ").to_numpy()

model = DPC(data)
model.calculate(0.6, 11, 4)
model.draw_decision_graph()
model.draw_cluster()
score = sklm.davies_bouldin_score(data, model.label.flatten())
print("DBI = {}".format(score))
```
